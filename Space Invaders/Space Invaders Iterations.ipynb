```python
'''
Completed Game!!!
'''

import sys
import pygame
import random

pygame.init()

clock = pygame.time.get_ticks()
prevclock = 0

score = 0
no_of_asteroids = 0

size = width, height = 1200, 1000
speed = [0, 0]
bulletspeed = -1.5
asteroidspeed = 1

screen = pygame.display.set_mode(size)
rocket = pygame.image.load("Rocketship.png")
asteroids = ['Asteroid1.png','Asteroid2.png','Asteroid3.png','Asteroid4.png']

rocket_rect = rocket.get_rect()
rocket_rect = rocket_rect.move(500,750)
direction = 0

asteroidlist = []
bulletlist = []


def fire():
    bulletlist.append([rocket_rect[0] + 50, rocket_rect[1]-50])

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT: sys.exit()
        '''
        Checking for keyboard input
        '''
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RIGHT:
                direction = 1
                speed[0] = 2.5
            elif event.key == pygame.K_LEFT:
                speed[0] = -2.5
                direction = -1
            if event.key == pygame.K_SPACE:
                print('FIRE!!')
                fire()
        if event.type == pygame.KEYUP:
            if direction == 1 and event.key == pygame.K_RIGHT:
                speed[0] = 0
            if direction == -1 and event.key == pygame.K_LEFT:
                speed[0] = 0
                
    if rocket_rect.left < 0 or rocket_rect.right > width:
        speed[0] = 0
        rocket_rect = rocket_rect.move([472-rocket_rect.left,0])

    '''
    Making asteroids
    '''
    if (clock - prevclock) > 2000:
        asteroidlist.append([asteroids[random.randint(0,3)],random.randint(100,width-100), 50])
        prevclock = clock
        no_of_asteroids += 1
    for item in asteroidlist:
        item[2] += asteroidspeed
        asteroid = pygame.image.load(item[0])
        asteroid_rect = asteroid.get_rect()
        asteroid_rect = asteroid_rect.move(item[1], item[2])
        screen.blit(asteroid, asteroid_rect)
        if item[2] > height+100:
            asteroidlist.remove(item)
            print(f'\t  \t Accuracy - {score/no_of_asteroids*100}')

    '''
    Moving bullets
    '''
    for item in bulletlist:
        item[1] += bulletspeed
        bullet = pygame.image.load("Bullet.png")
        bullet_rect = bullet.get_rect()
        bullet_rect = bullet_rect.move(item[0], item[1])
        screen.blit(bullet, bullet_rect)
        if item[1] < -100:
            bulletlist.remove(item)
            
    '''
    Checking if a bullet hit an asteroid
    '''
    for x in bulletlist:
        for y in asteroidlist:
            if abs(x[0] - y[1]) < 40 and abs(x[1] - y[2]) < 40:
                asteroidlist.remove(y)
                bulletlist.remove(x)
                score += 1
                print(f'Score - {score} \t Accuracy - {score/no_of_asteroids*100}')
                break
    
    clock = pygame.time.get_ticks()
    rocket_rect = rocket_rect.move(speed)
    screen.blit(rocket,rocket_rect)
    pygame.display.flip()
    screen.fill((0, 0, 0))

    '''
    Checking if rocket hit the asteroid
    '''
    for x in asteroidlist:
        if abs(rocket_rect[0] - x[1] + 80) < 100 and abs(rocket_rect[1] - x[2] + 100) < 80:
            print("The asteroid hit you!! \nBad luck you lost!!")
            pygame.quit()
```
